```python
import omnitool
from omnitool.literature_values import *
import barbershop

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import sys
```

```python
#Lets read in the Elsworth classified sample
sfile = '/home/oliver/PhD/Catalogues/RC_catalogues/Upgraded_Elsworth+17.txt'
cols = ['kic', 'CAT_OVERALL']

e17u = pd.read_csv(sfile,usecols=cols,sep='\s+')
e17u.rename(columns={'CAT_OVERALL':'stage'},inplace=True)
e17u.head(1)
# e17u.nu_max = e17u.nu_max.convert_objects(convert_numeric=True)
# e17u.delta_nu = e17u.delta_nu.convert_objects(convert_numeric=True)
```

```python
#Lets read in the APOKASC sample
sfile = '/home/oliver/PhD/Catalogues/APOKASC/APOKASC_cat_v3.6.5.txt'
cols= ['KEPLER_ID','TEFF_COR_POST','TEFF_COR_ERR',\
          'COR_NU_MAX','COR_NU_MAX_ERR',\
          'COR_DELTA_NU','COR_DELTA_NU_ERR','DR12_FE_H_ADOP_COR','K_MAG_2M']#\
#          'J_MAG_2M','J_MAG_ERR','K_MAG_2M','K_MAG_ERR','H_MAG_2M','H_MAG_ERR',\
#          'RA','DEC']

apokasc = pd.read_csv(sfile, usecols=cols, skiprows=644,sep='\s+')
apokasc.rename(columns={'KEPLER_ID':'kic','COR_NU_MAX':'numax','COR_DELTA_NU':'dnu',\
                       'COR_NU_MAX_ERR':'numax_err','COR_DELTA_NU_ERR':'dnu_err',\
                       'DR12_FE_H_ADOP_COR':'FeH',\
                       'TEFF_COR_POST':'Teff','TEFF_COR_ERR':'Teff_err'},inplace=True)
apokasc.head(1)
```

```python
#Lets combine the two into a single useful dataframe
df = pd.merge(e17u, apokasc, on='kic',how='outer')
#Kill anything not in the elsworth sample
df = df[~df.stage.isnull()]
```

```python
#Lets add a numerical stage indicator
numstage = np.ones(len(df))
numstage[df.stage=='RGB'] = 1
numstage[df.stage=='RC'] = 2
numstage[df.stage=='2CL'] = 3
numstage[df.stage=='U'] = 0
numstage[df.stage=='not_done'] = -1
df['numstage'] = numstage
```

```python
#Finally lets kill any of the weird -9999. values
print(len(df))
df = df[df.numax > 0.]
print(len(df))
df = df[df.dnu > 0.]
print(len(df))
df.head(2)
```

```python
#Lets run the scaling relations
sc = omnitool.scalings(df, df.numax, df.dnu, df. Teff,\
                       _numax_err = df.numax_err, _dnu_err = df.dnu_err,\
                       _Teff_err = df.Teff_err)

df['R'] = sc.get_radius()/Rsol
df['R_err'] = sc.get_radius_err()/Rsol
df['M'] = sc.get_mass()/Msol
df['M_err'] = sc.get_mass_err()/Msol
df['logg'] = sc.get_logg()
df['logg_err'] = sc.get_logg_err()
df['L'] = sc.get_luminosity()/Lsol
df['L_err'] = sc.get_luminosity_err()/Lsol
df['Mbol'] = sc.get_bolmag()
df['Mbol_err'] = sc.get_bolmag_err()
```

```python
#Lets get the bolometric corrections in K, J, and H
#Per Salaris et al. 1993, [M/H] = [Fe/H] + log(0.638*fa + 0.362)
#where fa is the alpha abundance. We will assume for now that fa = 1
#Therefore, [Fe/H] == [M/H]
Z = Zsol * 10 ** df.FeH.values
Teff = df.Teff.values
L = df.L.values
logg = df.logg.values

get_bc = omnitool.bolometric_correction(Teff,\
                                       logg,\
                                       L,\
                                       Z)
```

```python
#Lets get and apply the bolometric corrections
KBC = get_bc(band='Ks')
HBC = get_bc(band='H')
JBC = get_bc(band='J')
```

```python
#Calculate absolute magnitudes and appropriate error
df['MKs'] = df.Mbol - KBC
df['MH'] = df.Mbol - JBC
df['MJ'] = df.Mbol - HBC
df['absmag_err'] = np.sqrt(df.Mbol_err**2 + err_bc**2)
```
